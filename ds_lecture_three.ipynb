```python
plt.boxplot([whr_2018['Score'], whr_2019['Score']])
plt.xticks([1,2], ['2018', '2019'])
plt.show()
```

```python
whr_merged = whr_2018.merge(whr_2019, left_on='Country or region', right_on='Country or region', suffixes = ('_2018', '_2019'))
whr_merged
```

```python
scores = whr_merged[['Country or region', 'Score_2018', 'Score_2019']]
scores_max = scores.nlargest(15, 'Score_2018')
scores_min = scores.nsmallest(15, 'Score_2018')
```

```python
plt.figure(figsize=(20, 11))   # Defines size of the image in inches
plt.barh(scores_min['Country or region'], scores_min['Score_2018'], label='min 15')
plt.barh(scores_max['Country or region'], scores_max['Score_2018'],label='max 15')
plt.legend(loc='upper right',)
plt.gca().invert_yaxis()
plt.xlim(0, 8)
plt.ylim(0, 30)
```
